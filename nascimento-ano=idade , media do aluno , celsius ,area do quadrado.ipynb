```python
#Pedir para o usuário adicionar dois números e calcular a média deles.

num1 = int(input("digite um numero:"))
num2 = int(input("digite um numero:"))
num3 = int(input("digite outro numero:"))
num4 = int(input("digite outro numero:"))
print("media do usuario:",(num1+num2+num3)/num4)
```

```python
#Peça para o usuário calcular a idade de uma pessoa a partir do ano de nascimento.
nascimento = int(input("digite o ano de nascimento:"))
ano = int(input("digite o ano:"))
print("sua idade é:", nascimento-ano)
```

```python
 #Crie um programa que converta uma temperatura de Celsius para Fahrenheit.
 celsius = float(input("Digite a temperatura em celsius: "))
fahrenheit = (celsius * 1.8) + 32
print(f'A temperatura em celsius {celsius}°C corresponde à {fahrenheit}°F.')

```

```python
#Peça ao usuário para inserir a base e a altura de um retângulo e calcule a área.
Altura = int(input("digite a altura:"))
base = int (input("Digite a base"))
area = (base * Altura)/2
print("A area é:", area)
```
